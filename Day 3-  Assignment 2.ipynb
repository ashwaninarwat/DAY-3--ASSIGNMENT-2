#Question 1- use IF ELSE and ELIF to write a program in python for your report card.

#Question 2 - use for loop to print prime numbers btw 1 to 1000

```python
print('PRIME NUMBERS FROM 1 TO 1000')
for num in range (1,1001):
    if num>1:
        for i in range(2,num):
            if (num%i)==0:
                break
        else:
            print(num)
```

#Question 3- write a program for printing tables from 1 to 10 using for loop 

```python
for num in range(1,11):
    for t in range(1,11):
        print( num  ,"x" , t  ,"=" ,num*t, end ="\n")
```

#Write a program to print x prime numbers using while loop starting from 0 and take the input from the user

```python
num=int(input("enter the number : "))
c=2
while num!=0:
    for i in range(2,c):
        if c%i==0:
            break
    else:
        print(c)
        num-=1
    c+=1
```
